# WELCOME!

**Overview**

The data has been split into two groups:

- training set (train.csv)
- test set (test.csv)

**The training set** should be used to build your machine learning models. For the training set, we provide the outcome (also known as the ground truth) for each passenger. Your model will be based on features like passengers gender and class. You can also use feature engineering to create new features.

**The test set** should be used to see how well your model performs on unseen data. For the test set, we do not provide the ground truth for each passenger. It is your job to predict these outcomes. For each passenger in the test set, use the model you trained to predict whether or not they survived the sinking of the Titanic.

We also include gender_submission.csv, a set of predictions that assume all and only female passengers survive, as an example of what a submission file should look like.

**Data Dictionary**



**Variable Notes**

pclass: A proxy for socio-economic status (SES)

1st = Upper

2nd = Middle

3rd = Lower

age: Age is fractional if less than 1. If the age is estimated, is it in the form of xx.5

sibsp: The dataset defines family relations in this way...

Sibling = brother, sister, stepbrother, stepsister

Spouse = husband, wife (mistresses and fiancs were ignored)

parch: The dataset defines family relations in this way...

Parent = mother, father

Child = daughter, son, stepdaughter, stepson

Some children travelled only with a nanny, therefore parch=0 for them.

# Exploratory Data Analysis and Visualization

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
plt.rcParams["figure.figsize"] = (12,8)
import warnings
warnings.filterwarnings("ignore")
warnings.warn("this will not show")
```

```python
df = pd.read_csv("train.csv")
df.head()
```

```python
example_data =  df[["Pclass", "Sex", "Age", "SibSp", "Parch", "Fare", "Cabin", "Embarked", "Survived"]]
```

```python
example_data.info()
```

```python
example_data.isna().sum()
```

### CABIN

```python
example_data['Cabin'].value_counts()
```

***Cabin column has a lot of missing values, instead of dropping the missing values I wanted to identify for missing values as 0, and rest is 1***

```python
example_data['Cabin'] = example_data['Cabin'].replace(0, pd.np.nan)
```

```python
example_data['Cabin'] = example_data['Cabin'].notna().astype(int)
```

```python
example_data['Cabin'].value_counts()
```

### PARCH

```python
example_data
```

```python
example_data['Parch'].value_counts()
```

***For the values 5,3,6,4 are not enough data that can give the model good insight. Thats why I will remove those values*** 

```python
to_drop = example_data['Parch'].isin([5, 3, 4, 6])
example_data = example_data[~to_drop]
```

```python
example_data['Parch'].value_counts()
```

### EMBARKED 

```python
example_data['Embarked'].value_counts()
```

```python
example_data.isna().sum()
```

***Droping the missing values***

```python
example_data = example_data.dropna(subset=['Embarked'])
```

```python
example_data.isna().sum()
```

### AGE

```python
from fancyimpute import IterativeImputer
```

```python
mice_imputer = IterativeImputer()
```

```python
example_data['Age'] = mice_imputer.fit_transform(example_data[['Age']])
```

```python
example_data['Age'].describe().T
```

***Instead of dropping all the missing values in the Age column, I used the Iterative Imputer***

```python
df = example_data
```

## LABEL ENCODER

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['Sex'] = le.fit_transform(df['Sex'])
df['Embarked'] = le.fit_transform(df['Embarked'].astype(str))
```

```python
df
```

# TRAIN TEST SPLIT

```python
from sklearn.model_selection import train_test_split
```

```python
X = df[['Pclass', 'Sex', 'Age','SibSp', 'Parch', 'Fare', 'Cabin', 'Embarked']]
y = df['Survived']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

# AdaBoosting Classifier

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score, recall_score, precision_score, f1_score, roc_auc_score
from sklearn.model_selection import cross_val_score, cross_validate
```

```python
def eval_metric(model, X_train, y_train, X_test, y_test):
    y_train_pred = model.predict(X_train)
    y_pred = model.predict(X_test)
    
    print("Test_Set")
    print(confusion_matrix(y_test, y_pred))
    print(classification_report(y_test, y_pred))
    print()
    print("Train_Set")
    print(confusion_matrix(y_train, y_train_pred))
    print(classification_report(y_train, y_train_pred))
```

## VANILLA ADABOOST

```python
ada_model = AdaBoostClassifier(random_state = 42)
```

```python
ada_model.fit(X_train,y_train)
```

```python
eval_metric(ada_model, X_train, y_train, X_test, y_test)
```

```python
from sklearn.model_selection import cross_val_score, cross_validate

model = AdaBoostClassifier(random_state=42)

scores = cross_validate(model, X_train, y_train, scoring = ["accuracy", "precision", "recall", "f1"], cv = 10)
df_scores = pd.DataFrame(scores, index = range(1, 11))
df_scores.mean()[2:]
```

## ADABOOST GRIDSEARCH

```python
from sklearn.model_selection import GridSearchCV

model = AdaBoostClassifier(random_state=42)

param_grid = {"n_estimators": [20, 30, 100, 200], "learning_rate": [0.01, 0.1, 0.2, 0.5, 1.0]}
ada_grid_model = GridSearchCV(model, param_grid, cv = 5, scoring = "f1")
```

```python
ada_grid_model.fit(X_train, y_train)
```

```python
ada_grid_model.best_params_
```

```python
ada_grid_model.best_score_
```

```python
y_pred = ada_grid_model.predict(X_test)

ada_f1 = f1_score(y_test, y_pred)
ada_recall = recall_score(y_test, y_pred)
ada_auc = roc_auc_score(y_test, y_pred)
eval_metric(ada_grid_model, X_train, y_train, X_test, y_test)
```

from sklearn.metrics import precision_recall_curve, plot_precision_recall_curve, plot_roc_curve, roc_auc_score

```python
plot_roc_curve(ada_grid_model, X_test, y_test);
```

# GradientBoosting Classifier

## VANILLA GRADIENT BOOST

```python
from sklearn.ensemble import GradientBoostingClassifier
grad_model = GradientBoostingClassifier(random_state=42)
grad_model.fit(X_train, y_train)
```

```python
eval_metric(grad_model, X_train, y_train, X_test, y_test)
```

```python
model = GradientBoostingClassifier(random_state=42)

scores = cross_validate(model, X_train, y_train, scoring = ['accuracy', 'precision','recall',
                                                                   'f1', 'roc_auc'], cv = 10)
df_scores = pd.DataFrame(scores, index = range(1, 11))
df_scores.mean()[2:]
```

## GRIDSEARCH FOR GRADIENTBOOST

```python
param_grid = {"n_estimators":[100, 200, 300],
             "subsample":[0.5, 1], "max_features" : [None, 2, 3, 4], "learning_rate": [0.001, 0.01, 0.1], 'max_depth':[3,4,5,6]}
```

```python
gb_model = GradientBoostingClassifier(random_state=42)
```

```python
grid = GridSearchCV(gb_model, param_grid, scoring = "f1", verbose = 2, n_jobs= -1).fit(X_train, y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
y_pred = grid.predict(X_test)

gb_f1 = f1_score(y_test, y_pred)
gb_recall = recall_score(y_test, y_pred)
gb_auc = roc_auc_score(y_test, y_pred)

eval_metric(grid, X_train, y_train, X_test, y_test)
```

```python
plot_roc_curve(grid, X_test, y_test);
```

```python
param_grid = {"n_estimators":[50, 100, 200],'max_depth':[3,4,5], "learning_rate": [0.1, 0.2],
             "subsample":[0.5, 0.8, 1], "colsample_bytree":[0.5,0.7, 1]}

xgb_model = XGBClassifier(random_state=42)

xgb_grid = GridSearchCV(xgb_model, param_grid, scoring = "f1", verbose=2, n_jobs = -1).fit(X_train, y_train)
```

```python
xgb_grid.best_params_
```

```python
xgb_grid.best_score_
```

```python
y_pred = xgb_grid.predict(X_test)

xgb_f1 = f1_score(y_test, y_pred)
xgb_recall = recall_score(y_test, y_pred)
xgb_auc = roc_auc_score(y_test, y_pred)

eval_metric(xgb_grid, X_train, y_train, X_test, y_test)
```

```python
plot_roc_curve(xgb_grid, X_test, y_test);
```

# COMPARING MODELS

```python
compare = pd.DataFrame({"Model": ["AdaBoost",
                                 "GradientBoost", "XGBoost"],
                        "F1": [ada_f1, gb_f1, xgb_f1],
                        "Recall": [ada_recall, gb_recall, xgb_recall],
                        "ROC_AUC": [ada_auc, gb_auc, xgb_auc]})

def labels(ax):
    for p in ax.patches:
        width = p.get_width()                        # get bar length
        ax.text(width,                               # set the text at 1 unit right of the bar
                p.get_y() + p.get_height() / 2,      # get Y coordinate + X coordinate / 2
                '{:1.3f}'.format(width),             # set variable to display, 2 decimals
                ha = 'left',                         # horizontal alignment
                va = 'center')                       # vertical alignment
    
plt.figure(figsize=(14,10))
plt.subplot(311)
compare = compare.sort_values(by="F1", ascending=False)
ax=sns.barplot(x="F1", y="Model", data=compare, palette="Blues_d")
labels(ax)

plt.subplot(312)
compare = compare.sort_values(by="Recall", ascending=False)
ax=sns.barplot(x="Recall", y="Model", data=compare, palette="Blues_d")
labels(ax)

plt.subplot(313)
compare = compare.sort_values(by="ROC_AUC", ascending=False)
ax=sns.barplot(x="ROC_AUC", y="Model", data=compare, palette="Blues_d")
labels(ax)
plt.show()
```
